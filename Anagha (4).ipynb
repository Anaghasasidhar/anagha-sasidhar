# Kmeans clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,13):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(data)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

Plotting 

```python
plt.plot(range(1,13),wcss)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('WCSS score')
plt.show()
```

from elbow method optimum number of clusters is 3

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(data)
```

```python
from sklearn.metrics import silhouette_score
silhouette_sc = silhouette_score(data,y_kmeans)
print(silhouette_sc)
```

# silhouette score for Kmeans Clustering is 0.571

# AgglomerativeClustering

# Plotting Dendrogram to find the optimum number of clusters

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(data,method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Alcohol')
plt.ylabel('Euclidean Distance')
plt.show()
```

# From the dendogram we can divided the dataset into 2 clusters.

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2,affinity ='euclidean',linkage='ward')
y_hc = hc.fit_predict(data)
```

```python
y_hc
```

```python
#silhoutte score
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(data,y_hc)
print(silhouette_avg)
```

# silhouette score for AgglomerativeClustering is 0.658

# DB SCAN

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(n_neighbors=2)
nbrs = neigh.fit(data)
distances, indices = nbrs.kneighbors(data)
```

```python
distances, indices
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

From the map we can see that the optimum value of eps is 40.

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
```

```python
db = DBSCAN(eps=40,min_samples=5).fit(data)
labels = db.labels_
```

```python
db.labels_
```

```python
silhouette_avg = silhouette_score(data,db.labels_)
print(silhouette_avg)
```

# silhouette score for DBSCAN clustering is 0.5545
